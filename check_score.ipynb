```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def score60(d):
    return np.load(os.path.join('experiments', d, 'score_60.npy'))

def score90(d):
    return np.load(os.path.join('experiments', d, 'score_90.npy'))
```

```python
s60s_e120 = []
s90s_e120 = []
s60s_e125 = []
s90s_e125 = []
for b in range(4):
    epochs = 125
    try:
        s60 = score60(f"riab_cluster{b}_15_box_0635_epochs_{epochs}_steps_0_seq_20_batch_4000_g_6144_p_512_rf_003463636363636363_lr_00001_weight_decay_00001")
        s90 = score90(f"riab_cluster{b}_15_box_0635_epochs_{epochs}_steps_0_seq_20_batch_4000_g_6144_p_512_rf_003463636363636363_lr_00001_weight_decay_00001")
    except:
        s60 = score60(f"riab_cluster{b}_15_box_0635_epochs_200_steps_0_seq_20_batch_4000_g_6144_p_512_rf_003463636363636363_lr_00001_weight_decay_00001")
        s90 = score90(f"riab_cluster{b}_15_box_0635_epochs_200_steps_0_seq_20_batch_4000_g_6144_p_512_rf_003463636363636363_lr_00001_weight_decay_00001")
    s60s_e125.append(s60)
    s90s_e125.append(s90)

    epochs = 120
    s60 = score60(f"riab_cluster{b}_15_box_0635_epochs_{epochs}_steps_0_seq_20_batch_4000_g_6144_p_512_rf_003463636363636363_lr_00001_weight_decay_00001")
    s90 = score90(f"riab_cluster{b}_15_box_0635_epochs_{epochs}_steps_0_seq_20_batch_4000_g_6144_p_512_rf_003463636363636363_lr_00001_weight_decay_00001")
    s60s_e120.append(s60)
    s90s_e120.append(s90)

plt.figure(figsize=(6, 5))
plt.plot(np.median(s60s_e120, axis=1), lw=5, label='120 epochs')
plt.plot(np.median(s60s_e125, axis=1), lw=5, label='125 epochs')
plt.ylim([0, 0.3])
plt.xticks(range(4), [f'Cluster {i}' for i in range(4)])
plt.title('Score@60')
plt.legend()
plt.show()

plt.figure(figsize=(6, 5))
plt.plot(np.median(s90s_e120, axis=1), lw=5, label='120 epochs')
plt.plot(np.median(s90s_e125, axis=1), lw=5, label='125 epochs')
plt.ylim([0, 0.6])
plt.xticks(range(4), [f'Cluster {i}' for i in range(4)])
plt.title('Score@90')
plt.legend()
plt.show()
```
